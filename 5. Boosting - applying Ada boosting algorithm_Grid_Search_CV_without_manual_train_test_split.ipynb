```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
Indep = dataset [['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep = dataset [['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test = train_test_split(Indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble  import AdaBoostRegressor

grid_params = {'n_estimators':[10,20,30,40,50,100],'learning_rate' :[1.0,2.0,3.0,4.0],'random_state' :[0,1,2,3,4,5]}

grid = GridSearchCV(AdaBoostRegressor(),grid_params,refit=True,verbose=3,n_jobs=-1)
               
grid.fit(Indep,dep)
```

```python
y_pred = grid.predict(Indep)
```

```python
from sklearn.metrics import r2_score
result = r2_score(dep,y_pred)
result
```

```python
best_param = grid.best_params_
best_score = grid.best_score_
best_rank  = grid.best_index_
(best_rank,best_param,best_score)
```

```python
table = grid.cv_results_
```

```python
final_table = pd.DataFrame.from_dict(table)    
final_table
```
