```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
Independent = dataset [['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset [['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test = train_test_split(Independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
import xgboost as xgb

grid_params = {'n_estimators':[10,20,30,40,50,100],'learning_rate' :[1.0,2.0,3.0,4.0],'random_state' :[0,1,2,3,4,5]}

grid = GridSearchCV(xgb.XGBRegressor(),grid_params,refit=True,verbose=3,n_jobs=-1)

grid.fit(Independent,dependent)
```

```python
y_pred = grid.predict(Independent)
y_pred
```

```python
from sklearn.metrics import r2_score
result = r2_score(dependent,y_pred)
result
```

```python
result = grid.cv_results_

print("Best param",grid.best_params_,grid.best_score_)
```

```python
table = pd.DataFrame.from_dict(result)      
table
```
