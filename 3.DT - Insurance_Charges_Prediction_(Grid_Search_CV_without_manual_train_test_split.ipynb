```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
Independent = dataset [['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset [['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test = train_test_split(Independent,dependent,train_size=0.70,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

grid_params = {'criterion':['mse','mae','friedman_mse'],'max_features': ['auto','sqrt','log2'],'splitter':['best','random']}

grid = GridSearchCV(DecisionTreeRegressor(), grid_params, refit=True, verbose=3, n_jobs=-1 )

grid.fit(Independent,dependent)
```

```python
#from sklearn import tree
#import matplotlib.pyplot as plt
#tree.plot_tree(DT)
#plt.show()
```

```python
y_pred = grid.predict(Independent)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(dependent,y_pred)
R2
```

```python
Result = grid.cv_results_
```

```python
best = grid.best_params_
best_score = grid.best_score_
print ("Best performing Params",best,best_score)
```

```python
final_Result = pd.DataFrame.from_dict(Result)
final_Result 
```

```python
Age = int(input("Enter the age:"))
bmi = float(input("Enter the bmi:"))
children = int(input("Enter the no. of children:"))
sex = int(input("Enter the Sex:"))
smoker = int(input("Enter the no. of smoker:"))

final_output = grid.predict([[Age,bmi,children,sex,smoker]])
print("The final predicted insurance charges{}:".format(final_output))
```
