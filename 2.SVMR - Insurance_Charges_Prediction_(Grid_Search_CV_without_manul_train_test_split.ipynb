```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
Independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

grid_params = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000]}

grid = GridSearchCV (SVR(),grid_params,refit=True,verbose=3,n_jobs=-1)
grid.fit(Independent,dependent)
```

```python
y_pred = grid.predict(Independent)
```

```python
from sklearn.metrics import r2_score
R2_Result = r2_score(y_pred,dependent)
R2_Result
```

```python
get_result= grid.cv_results_
```

```python
g=grid.best_params_
print("The best performing hyper playable params are :{}".format(g),grid.best_score_)
```

```python
final_result = pd.DataFrame.from_dict(get_result)
final_result
```
