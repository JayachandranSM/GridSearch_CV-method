```python
import pandas as pd
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset = dataset.drop("User ID",axis=1)
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
Indep = dataset [['Age', 'EstimatedSalary', 'Gender_Male']]
dep = dataset [['Purchased']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test= train_test_split(Indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

grid_params = {'n_estimators':[10,20,30,40,50],'max_features' : ['log2','sqrt','auto']}

grid = GridSearchCV(RandomForestClassifier(),grid_params,refit=True,verbose=3,n_jobs=-1)

grid.fit(Indep,dep)
```

```python
y_pred = grid.predict(Indep)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(dep,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
report = classification_report(dep,y_pred)
print(report)
```

```python
result = grid.cv_results_

print("The best performing param:",grid.best_estimator_,grid.best_params_,grid.best_score_)
```

```python
table = pd.DataFrame.from_dict(result)
table
```
